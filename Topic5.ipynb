```python
import os
import re
import json
import cv2
import numpy as np
import pandas as pd
import pytesseract
import easyocr

# ================= CONFIG =================
BASE_DIR = "/content/drive/MyDrive/yolov9"
IMAGE_PATH = f"{BASE_DIR}/chart_11.jpg"

# YOLO detect outputs
BAR_LABEL_PATH = f"{BASE_DIR}/runs/detect/exp25/labels/chart_11.txt"
TEXT_LABEL_PATH = f"{BASE_DIR}/runs/detect/exp26/labels/chart_11.txt"

SAVE_DIR = f"{BASE_DIR}/runs/detect/final"
os.makedirs(SAVE_DIR, exist_ok=True)

# Class IDs
TITLE_CLASS_ID = 0
X_LABEL_CLASS_ID = 1
Y_LABEL_CLASS_ID = 2

# ================= STEP 1: READ IMAGE =================
img = cv2.imread(IMAGE_PATH)
if img is None:
    raise FileNotFoundError("Không đọc được ảnh")

h_img, w_img = img.shape[:2]

# ================= STEP 2: READ BAR DETECTION =================
bars = []
with open(BAR_LABEL_PATH) as f:
    for i, line in enumerate(f):
        parts = list(map(float, line.strip().split()))
        if len(parts) < 5:
            continue
        cls, x, y, bw, bh = parts[:5]
        conf = parts[5] if len(parts) == 6 else 1.0

        pixel_h = bh * h_img
        bars.append({
            "id": i + 1,
            "x_center": x * w_img,
            "y_center": y * h_img,
            "height_px": pixel_h,
            "confidence": round(conf, 3)
        })

if not bars:
    raise ValueError("Không phát hiện bar")

bars = sorted(bars, key=lambda b: b["x_center"])

# ================= STEP 3: DETECT Y-AXIS =================
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, minLineLength=50, maxLineGap=10)

y_axis = None
if lines is not None:
    for l in lines:
        x1, y1, x2, y2 = l[0]
        if abs(x1 - x2) < 5 and x1 < w_img * 0.3:
            y_axis = (x1, y1, x2, y2)
            break

if y_axis is None:
    raise ValueError("Không tìm thấy trục Y")

y_top = min(y_axis[1], y_axis[3])
y_bottom = max(y_axis[1], y_axis[3])

# Crop Y-axis for OCR
pad_x, pad_y = 60, 10
x1 = max(0, y_axis[0] - pad_x)
x2 = min(w_img, y_axis[0] + pad_x)
y1 = max(0, y_top - pad_y)
y2 = min(h_img, y_bottom + pad_y)

y_crop = img[y1:y2, x1:x2]
y_crop_path = f"{SAVE_DIR}/y_axis_crop.png"
cv2.imwrite(y_crop_path, y_crop)

# ================= STEP 4: OCR Y-AXIS VALUES =================
big = cv2.resize(y_crop, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)
gray_big = cv2.cvtColor(big, cv2.COLOR_BGR2GRAY)
th = cv2.adaptiveThreshold(gray_big, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                           cv2.THRESH_BINARY, 25, 10)

ocr_cfg = r"--oem 3 --psm 6 -c tessedit_char_whitelist=0123456789"
text_ocr = pytesseract.image_to_string(th, config=ocr_cfg)
numbers = list(map(int, re.findall(r"\d+", text_ocr)))

y_max_val = max(numbers) if numbers else 100

# ================= STEP 5: COMPUTE REAL BAR VALUES =================
bar_results = []
for b in bars:
    y_top_bar = b["y_center"] - b["height_px"] / 2
    real_val = ((y_bottom - y_top_bar) / (y_bottom - y_top)) * y_max_val

    bar_results.append({
        "x_center": round(b["x_center"], 2),
        "real_value": round(real_val, 2),
        "confidence": b["confidence"]
    })

with open(f"{SAVE_DIR}/bar_real_values.json", "w", encoding="utf-8") as f:
    json.dump(bar_results, f, indent=4, ensure_ascii=False)

# ================= STEP 6: OCR TEXT (TITLE, AXES, TICKS) =================
reader = easyocr.Reader(['en'])
ocr_results = []
category_bbox = None

expand_ratio = 0.5  # mở rộng bbox trước khi OCR

with open(TEXT_LABEL_PATH) as f:
    for line in f:
        cls, x, y, bw, bh, conf = map(float, line.strip().split())

        x1 = int((x - bw/2) * w_img)
        y1 = int((y - bh/2) * h_img)
        x2 = int((x + bw/2) * w_img)
        y2 = int((y + bh/2) * h_img)

        # ---- EXPAND BBOX ----
        dw = int(bw * w_img * expand_ratio)
        dh = int(bh * h_img * expand_ratio)
        x1 = max(0, x1 - dw)
        y1 = max(0, y1 - dh)
        x2 = min(w_img, x2 + dw)
        y2 = min(h_img, y2 + dh)

        crop = img[y1:y2, x1:x2]
        if crop.size == 0:
            continue

        # chữ dọc → xoay
        if (y2 - y1) > 1.5 * (x2 - x1):
            crop = cv2.rotate(crop, cv2.ROTATE_90_CLOCKWISE)

        text = " ".join([r[1] for r in reader.readtext(crop)])

        ocr_results.append({
            "class_id": int(cls),
            "text": text,
            "bbox": (x1, y1, x2, y2)
        })

        if int(cls) == X_LABEL_CLASS_ID:
            category_bbox = (x1, y1, x2, y2)

# -----------------------------
# X-TICKS OCR
# -----------------------------
x_ticks = []
if category_bbox:
    x1, y1, x2, y2 = category_bbox

    margin = 5
    ocr_height = 120

    tx1 = max(0, x1 - 30)
    ty1 = max(0, y1 - ocr_height - margin)
    tx2 = min(w_img, x2 + 60)
    ty2 = max(0, y1 - margin)

    tick_crop = img[ty1:ty2, tx1:tx2]

    if tick_crop.size > 0:
        tick_ocr = reader.readtext(tick_crop)
        x_ticks = [r[1] for r in tick_ocr]

# ================= STEP 7: BUILD FINAL OUTPUT =================
chart_title = next((o['text'] for o in ocr_results if o['class_id']==TITLE_CLASS_ID), "")
x_label = next((o['text'] for o in ocr_results if o['class_id']==X_LABEL_CLASS_ID), "")
y_label = next((o['text'] for o in ocr_results if o['class_id']==Y_LABEL_CLASS_ID), "")

pairs = []
for tick, bar in zip(x_ticks, bar_results):
    pairs.append((tick, bar['real_value']))

output_lines = []
output_lines.append(f"Chart Title: {chart_title}")
output_lines.append(f"Y-Label: {y_label}")
output_lines.append(f"X-Label: {x_label}")
output_lines.append("DataTable:")
output_lines.append("\tX:Y")
for t, v in pairs:
    output_lines.append(f"\t{t}: {v}")

final_txt = "\n".join(output_lines)

with open(f"{SAVE_DIR}/final_output.txt", "w", encoding="utf-8") as f:
    f.write(final_txt)

print("=== DONE ===")
print(final_txt)



```
